```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('datasets/stocks.csv')
```

```python
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.head()
```

```python
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.05,0.65,0.5,0.3])

ax1.set_title('IBM Vs ORCL')
ax1.plot(data['Date'], data['ORCL'], color='green')
ax2.plot(data['Date'], data['IBM'], color= 'blue')
plt.show()
```

```python
fig = plt.figure(figsize=(16,10))

fig.suptitle("Stock price comparison")

ax1 = fig.add_subplot(331)
ax1.set_title('AAPL')
ax1.plot(data['Date'],data['AAPL'], color = 'm')

ax2 = fig.add_subplot(332)
ax2.set_title('ADBE')
ax2.plot(data['Date'], data['ADBE'], color = 'blue')

ax3 = fig.add_subplot(333)
ax3.set_title('CVX')
ax3.plot(data['Date'], data['CVX'], color = 'green')

ax4 = fig.add_subplot(334)
ax4.set_title('GOOG')
ax4.plot(data['Date'], data['GOOG'], color = 'purple')

ax5 = fig.add_subplot(335)
ax5.set_title('MDLZ')
ax5.plot(data['Date'], data['MDLZ'], color = 'teal')

ax6 = fig.add_subplot(336)
ax6.set_title('MSFT')
ax6.plot(data['Date'], data['MSFT'], color = 'cyan')

ax7 = fig.add_subplot(337)
ax7.set_title('NFLX')
ax7.plot(data['Date'], data['NFLX'], color = 'red')

ax8 = fig.add_subplot(338)
ax8.set_title('SBUX')
ax8.plot(data['Date'], data['SBUX'], color = 'orange')

plt.show()
```
